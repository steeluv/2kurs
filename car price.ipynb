## С помощью данной функции мы удалили все лишние запятые в значениях столбцов "Old Price" и "New Price"

***

```python
def to_int (column ):
    car[column] = car[column].astype(str).replace(r'\D', '', regex=True)
    for i in range(len(car[column])):
        car[column][i] = car[column][i].replace("," , "")
```

```python
to_int("Old Price")
to_int("New Price")
```

```python
car.loc[:, "Old Price"] = car["Old Price"].astype(int)
car.loc[:, "New Price"] = car["New Price"].astype(int)
```

# Выведем процентное соотношение количества машин, выпущенных в определенных годах

```python
data = [360, 4, 2676, 433, 31]
labels = ["2023","2019", "2022", "2021", "2020"]
colors = sns.color_palette('bright')[4:9]
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

## Можем сделать следующий вывод:
