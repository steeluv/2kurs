```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv(r"C:\\Users\\Steluv\\Downloads\\Titanic.csv")
```

```python
titanic.drop(['Name','Ticket','Cabin'], axis = 1)
```

```python
titanic['Age'].fillna(titanic['Age'].mean())
```

```python
sns.lineplot(data = titanic, x = "PassengerId", y = "Age")
```

```python
data = [644, 168, 77]
labels = ['S', 'C', 'Q']
colors = sns.color_palette('bright')[6:10]
plt.pie(data, labels = labels, colors = colors, autopct ='%0f%%')
plt.show()
```
